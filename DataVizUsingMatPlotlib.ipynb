```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[50,51,52,53,54,45,56]
```

```python
plt.plot(x,y,color="green",linestyle="solid")
plt.title("weather")
plt.xlabel("day")
plt.ylabel("temp")
```

```python
# format string
plt.plot(x,y,"r+--",alpha=0.25)
```

```python
# axes labels,legends,grid
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,52]
min_t=[40,42,40,44,33,35,37]
avg_t=[44,48,48,46,42,41,44]
```

```python
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.xlabel("days")
plt.ylabel("temp")
plt.title("weather")
plt.legend(loc="best",shadow=True)
plt.grid()
```
